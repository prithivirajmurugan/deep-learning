## Thermodynamic process

* **Adiabatic process** - No heat transfer into or out of the system occurs
* **Isochoric process** - No chnage in volume occurs
* **Isobaric process** - No change in pressure occurs
* **Isothermal process** - No change in temperature occurs

## Four-Stroke Cycle Engine

Takes four strokes to complete the combustion cycle

1. Intake stroke
2. Compression stroke
3. Power stroke
4. Exhaust stroke

image.png

## Operation:

**Intake stroke** : Piston move from Top-Dead-Centre toward Bottom-Dead-Centre, which increases the volume of the cylinder,creating a low-pressure and the outside air or air/fuel mixture is pulled in the cylinder.

**Compression stroke** : The piston moving from its position of TDC to the BDC of the cylinder, which compresses the air inside the cylinder, decrease in the volume increases the cyclinder pressure, which causes the increase the temperature of the working fluid

**Power stroke** : with the increase of temperature in working fluid, fuel is injected and ignited, which inturns generates heat and causes the working fluid to expand creating a high pressure, pushing the piston towards TDC, this is called as Expansion Stroke or Power stroke as this stroke generates the over-all power output

**Exhaust stroke** - At the end of the Power stroke, the piston returns back to the BDC from the TDC, which pushes the exhaust gases (combustion byproduct), outside of the cylinder and becomes ready for the next cycle.

# Thermal efficiency

Ratio between the useful work done by the engine to the thermal energy supplied to it.

$\eta_{th} = Work Output / Thermal Energy Supplied to the Control Volume $

\eta_{th} is cannot be equal to 100% as the convertion of thermal energy to work will include energy losses

# Engine Performance Parameters

## Power Output 

Maximum torque at the given engine speed

$ Power Output = T*N / Constant $

Power is the rate at which the work is supplied

The Power is usually measured using a dynamo-meter

$ T = F * r $

Work done for one revolution

$ Work = 2 * \pi * r * F $

$ Work = 2 * \pi * T $

Work done per unit time

$ Work = 2 * \pi * T * N $

Work done per minute = Power

$ Power = 2 * \pi * T * N $




## Indicated Power

Brake power is the power output result of the application of engine to a load where as indicated power is the power as the result of the gas compression and expansion action in the cylinder

$ Indicated Power = Brake Power + Pumping Power + Friction Power $


## Indicated Mean Effective Pressure

## Break Mean Effective Pressure

Portion of the bulk pressure exerted on the surface of the piston ($ imep $), that produces useful power

It represents the net work after subtracting friction, pumping, and other parasitic losses (the largest of
these losses being that of friction)

## Specific Fuel Consumption

How much fuel the engine consumes in the process of producing an output of one horsepower or kilowatt

$ bsfc = Fuel Consumption per Unit Time / Brake Power Output $ 

$ isfc = Fuel Consumption per Unit Time / indicated Power Output $ 

Engine Specific Volume

$ Specific Weight = Engine Weight / Rated Power $
